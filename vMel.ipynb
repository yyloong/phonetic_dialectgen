```python
# 可视化 mel-spectrogram
import numpy as np
import matplotlib.pyplot as plt 
import torch
import librosa
import librosa.display
```

```python
def plot_mel_spectrogram(mel_spectrogram, sample_rate=22050):
    """
    可视化 mel-spectrogram
    :param mel_spectrogram: 形状为 [B, T, C] 的 mel-spectrogram
    :param sample_rate: 音频采样率
    """
    # 取第一个样本进行可视化
    mel_spectrogram = mel_spectrogram[0]
    
    # 绘制 mel-spectrogram
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mel_spectrogram.T, sr=sample_rate, x_axis='time', y_axis='mel', cmap='viridis')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel-Spectrogram')
    plt.tight_layout()
    plt.show()
```

```python
# 导入 mel-spectrogram (.npy 文件)
mel_spectrogram = torch.load('mel_output.pth')
print(f'mel_spectrogram shape: {mel_spectrogram.shape}')
mel_spectrogram.shape, mel_spectrogram.min(), mel_spectrogram.max(), mel_spectrogram.mean(), mel_spectrogram.std()
mel_spectrogram = mel_spectrogram.detach().cpu().numpy()
# 可视化
plot_mel_spectrogram(mel_spectrogram)
```

```python
test_mel = torch.load("../backup/edge-mel/124.pt")
print(test_mel.shape)
test_mel = test_mel.cpu().numpy().T
test_mel = test_mel.reshape(1, test_mel.shape[0], test_mel.shape[1])            # 添加 batch 维度
test_mel.shape, test_mel.min(), test_mel.max(), test_mel.mean(), test_mel.std()
```
